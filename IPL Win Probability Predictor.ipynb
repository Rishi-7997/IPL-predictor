# Feature Engineering

```python
# grouping the 1st innings,2nd innings score in a particular matchid
# lets say match id = 1,so inning 1 score = 207,inning 2 score = 172,in that way

totalrun_df = deliveries.groupby(['match_id','inning']).sum()['total_runs'].reset_index()

totalrun_df
```

```python
# capturing only the first innings,as we will be predicting for the second innnigs
totalrun_df = totalrun_df.loc[totalrun_df['inning'] == 1]
totalrun_df.loc[:, 'total_runs'] = totalrun_df['total_runs'] + 1
```

```python
matches.columns
```

```python
'''
Merging the total first innings score df with the matches df,
where left side merging is done on "id" column of the matches
and right side merging is done on "match_id" column of the totalrun_df

'''

match_df = matches.merge(totalrun_df[['match_id','total_runs']],
                       left_on='id',right_on='match_id')

match_df
```

```python
match_df['team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
# replacing the Delhi Daredevils with Delhi Capitals

match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')


# replacing the Deccan Chargers with Sunrises Hyderabad

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

```

```python
# will consider only frequently occuring teams,
# which are mentioned in the teams list

match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]

match_df['team1'].unique()
```

```python
# ignoring the rows which were DL method

match_df = match_df[match_df['dl_applied']==0]

# considering the match_id,city,winner and total runs

match_df = match_df[['match_id','city','winner','total_runs']]

match_df
```

```python
deliveries.head(3)
```

```python
# merging matchdf with delevieries on match_id

delivery_df = match_df.merge(deliveries,on='match_id')

delivery_df.head(5)
```

```python
delivery_df.columns
```

```python
# considering the 2nd innings 

delivery_df = delivery_df[delivery_df['inning'] == 2]
delivery_df.head()
```

```python
delivery_df.shape
```

```python
'''
So by observation we can observe that in the matchdf we had taken
firstinnings total runs,right and in the second case,that is in the 
delivery dataframe we considered second inning runs,as our main aim
is to find the probability of either teams to win or loose,we need 
current runs and runrate,so for current runs,we can apply groupby
on matchid and take the cummulative sum wrt total_runs_y,now,basically
totalruns was present in matchdf as well as deliveries_df,but as we merged
both the dataframes,it resulted in total_runs_x,and total_runs_y,
so total_runs_x is the first innings runs and total_runs_y are the second
innings runs,ball by ball,by applying cummulative sum,this becomes 
current score


total_runs_y gives the run scored after each ball,so in the second innings,
we want to get the total second innings runs,so we will groupby match id
and will apply the cummulative sum


'''

# current score of particular match
delivery_df['total_runs_y'] = pd.to_numeric(delivery_df['total_runs_y'], errors='coerce')

# Apply cumulative sum within each group
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()

delivery_df.head()
```

```python
# runs left 

delivery_df['runs_left'] = delivery_df['total_runs_x']-delivery_df['current_score']

delivery_df.head()
```

```python
'''
if one ball is played,then balls left = 120-1 = 119........(1)
if two balls are played,then balls left = 120-2 = 118

so similarly if over=1,over has 6 balls right,so 1*6 = 6
now,ball = 1,so 6+1 = 7,now 126-7 = 119,which is same as (1)

so we'll use balls_left = 126-(over*6+current_ball)

'''

# balls left


delivery_df['balls_left'] = 126-(delivery_df['over']*6+delivery_df['ball'])

delivery_df
```

```python
list(delivery_df['player_dismissed'].unique())[:2]
```

```python

# filling nan values with "0"

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")

# now we will convert this player_dismissed col into a boolean col
# if the player is not dismissed then it's 0 else its 1

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x==0 else 1)

# now we will convert this player_dismissed col into a boolean col

delivery_df['player_dismissed'].unique()
```

```python
# Convert 'player_dismissed' to binary: 1 if a player is dismissed, 0 otherwise
delivery_df['wicket_fallen'] = delivery_df['player_dismissed'].notnull().astype(int)

# Apply cumulative sum within each group
delivery_df['wickets_fallen'] = delivery_df.groupby('match_id')['wicket_fallen'].cumsum()

# Calculate remaining wickets
delivery_df['wickets_remaining'] = 10 - delivery_df['wickets_fallen']

delivery_df.head()
```

```python
# current run rate

delivery_df['cur_run_rate'] = (delivery_df['current_score']*6)/(120-delivery_df['balls_left'])

# required run rate

delivery_df['req_run_rate'] = (delivery_df['runs_left']*6)/(delivery_df['balls_left'])


delivery_df.head(3)
```

```python
def resultfun(row):
    
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(resultfun,axis=1)
delivery_df.head()
```

```python
numvars = [x for x in final_df.columns if final_df[x].dtypes != 'O']
catvars = [x for x in final_df.columns if final_df[x].dtypes == 'O']
```

```python
def plot_distribution(final_df, columns, bins=10):
    for col in columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(final_df[col], bins=bins, kde=True)
        plt.title(f'Distribution of {col.replace("_", " ").title()}')
        plt.xlabel(col.replace("_", " ").title())
        plt.ylabel('Frequency')
        plt.show()

def plot_scatter(final_df, x_y_pairs):
    for x, y in x_y_pairs:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=x, y=y, data=final_df)
        plt.title(f'{x.replace("_", " ").title()} vs {y.replace("_", " ").title()}')
        plt.xlabel(x.replace("_", " ").title())
        plt.ylabel(y.replace("_", " ").title())
        plt.show()

def plot_count(final_df, columns):
    for col in columns:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=col, data=final_df)
        plt.title(f'Count of {col.replace("_", " ").title()}')
        plt.xlabel(col.replace("_", " ").title())
        plt.ylabel('Count')
        plt.xticks(rotation=90)
        plt.show()
```

```python
# Pairs of columns for scatter plots (only numeric columns)
scatter_pairs = [
    ('runs_left', 'balls_left'),
    ('cur_run_rate', 'req_run_rate')
]
```

```python
# Distribution plots for numerical variables
plot_distribution(final_df, numvars, bins=5)

# Scatter plots for numerical variables
plot_scatter(final_df, scatter_pairs)

# Count plots for categorical variables
plot_count(final_df, catvars)

# Correlation Heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = final_df[numvars].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Pairplot for Multivariate Analysis
sns.pairplot(final_df[numvars])
plt.suptitle('Pairplot of Key Features', y=1.02)
plt.show()
```

```python
data = final_df.copy()

test = data['result']

train = data.drop(['result'],axis = 1)

train.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(train,test,test_size=0.2,random_state=1)

X_train.shape,X_test.shape
```

```python
X_train.columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, recall_score, accuracy_score
```

```python
# batting team,bowling team and city are categorical columns
# they will be converted to numeric using onehot encoder

cf = ColumnTransformer([
    ('trf',OneHotEncoder(drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
# Selecting Logistic Regression
# creating the pipeline

pipe = Pipeline(steps=[
    ('step1',cf),
    ('step2',LogisticRegression(solver='liblinear'))
])

# Fit the model on training data
pipe.fit(X_train, y_train)

# Accuracy Score on Test Set
y_pred = pipe.predict(X_test)
print(f"Accuracy on Test Set: {accuracy_score(y_test, y_pred)* 100:.2f}")

# Cross-Validation for Accuracy
accuracy_scores = cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy')
print(f"Cross-Validation Accuracy: {accuracy_scores.mean()* 100:.2f}")

# Cross-Validation for Recall
recall_scorer = make_scorer(recall_score, average='macro') 
recall_scores = cross_val_score(pipe, X_train, y_train, cv=5, scoring=recall_scorer)
print(f"Cross-Validation Recall: {recall_scores.mean()* 100:.2f}")
```

```python
pipe.predict_proba(X_test)[10]
```

There is a 30% probability that the instance (in this case, the 11th example in the test set) belongs to the loss class (class 0).

There is a 69% probability that the instance belongs to the win class (class 1).

```python
# Selecting Random Forest Classifier
# Define the pipeline

pipe2 = Pipeline(steps=[
    ('step1', cf),
    ('step2', RandomForestClassifier())
])

# Fit the model on training data
pipe2.fit(X_train, y_train)

# Accuracy Score on Test Set
y_pred = pipe2.predict(X_test)
print(f"Accuracy on Test Set: {accuracy_score(y_test, y_pred)* 100:.2f}")

# Cross-Validation for Accuracy
accuracy_scores = cross_val_score(pipe2, X_train, y_train, cv=5, scoring='accuracy')
print(f"Cross-Validation Accuracy: {accuracy_scores.mean()* 100:.2f}")

# Cross-Validation for Recall
recall_scorer = make_scorer(recall_score, average='macro')
recall_scores = cross_val_score(pipe2, X_train, y_train, cv=5, scoring=recall_scorer)
print(f"Cross-Validation Recall: {recall_scores.mean()* 100:.2f}")
```

```python
# oth index is loose prob and 1st index is win prob

pipe2.predict_proba(X_test)[10]
```

There is a 0% probability that the instance (in this case, the 11th example in the test set) belongs to the loss class (class 0).

There is a 100% probability that the instance belongs to the win class (class 1).
